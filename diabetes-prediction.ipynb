### Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import joblib
import warnings
warnings.filterwarnings("ignore")
sns.set_style("whitegrid")
```

### Data visualization

```python
plt.figure(figsize=(12, 5))

# Bar Chart
plt.subplot(1, 2, 1)

outcome_counts = diabetes_data['Outcome'].value_counts()

plt.bar(
    ['Non-Diabetic', 'Diabetic'],
    outcome_counts.values,
    color=['#2ecc71', '#e74c3c']
)

plt.title('Distribution of Diabetes Outcomes', fontsize=14, fontweight='bold')
plt.ylabel('Count')

# Pie Chart
plt.subplot(1, 2, 2)

plt.pie(
    outcome_counts.values,
    labels=['Non-Diabetic', 'Diabetic'],
    autopct='%1.1f%%',
    colors=['#2ecc71', '#e74c3c'],
    startangle=90
)

plt.title('Percentage Distribution', fontsize=14, fontweight='bold')
```

```python
# Figure Size
plt.figure(figsize=(10, 8))

# Correlation Matrix
correlation = diabetes_data.corr()

# Heatmap
sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=1,
    fmt='.2f'
)

# Title
plt.title(
    'Feature Correlation Heatmap',
    fontsize=14,
    pad=20,
    fontweight='bold'
)

plt.show()
```

```python
# Create 3x3 subplot
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# Select all feature columns except Outcome
features = diabetes_data.columns[:-1]

# Loop through each feature
for idx, feature in enumerate(features):

    row = idx // 3
    col = idx % 3

    # Plot histogram for each Outcome
    for outcome in [0, 1]:

        data = diabetes_data[diabetes_data['Outcome'] == outcome][feature]

        axes[row, col].hist(
            data,
            bins=20,
            alpha=0.6,
            label=f'Outcome {outcome}',
            color='#2ecc71' if outcome == 0 else '#e74c3c'
        )

    axes[row, col].set_title(feature, fontweight='bold')
    axes[row, col].set_xlabel('Value')
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].legend()
plt.show()
```

### Data Preprocessing

```python
X = diabetes_data.drop('Outcome', axis=1)
y = diabetes_data['Outcome']

print(X)
print(y)
```

### Standarization

```python
scaler = StandardScaler()
x = scaler.fit_transform(X)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

print(X.shape, x_train.shape, x_test.shape)
```

### Build the model

```python
classifier = svm.SVC(kernel='linear', probability=True)
classifier.fit(x_train, y_train)
```

```python
x_train_predictions = classifier.predict(x_train)
training_accuracy = accuracy_score(x_train_predictions, y_train)
training_accuracy
```

```python
x_test_predictions = classifier.predict(x_test)
test_accuracy = accuracy_score(x_test_predictions, y_test)
test_accuracy
```

```python
precision = precision_score(y_test, x_test_predictions)
recall = recall_score(y_test, x_test_predictions)

f1 = f1_score(y_test, x_test_predictions)
print(precision, recall, f1)
```

```python
cm_svm = confusion_matrix(y_test, x_test_predictions)
cm_svm
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', square= True,
            xticklabels=['Non-Diabetic', 'Diabetic'],
            yticklabels=['Non-Diabetic', 'Diabetic'])
plt.title('Confusion Matrix for SVM Classifier', fontsize=14, fontweight='bold')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

### Random Forest Model 

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=2)
rf_classifier.fit(x_train, y_train)
```

```python
x_train_rf_predictions = rf_classifier.predict(x_train)
```

```python
training_accuracy_rf = accuracy_score(x_train_rf_predictions, y_train)
training_accuracy_rf
```

```python
x_test_rf_predictions = rf_classifier.predict(x_test)
test_accuracy_rf = accuracy_score(x_test_rf_predictions, y_test)
test_accuracy_rf
```

```python
precision_rf = precision_score(y_test, x_test_predictions)
```

```python
recall_rf = recall_score(y_test, x_test_predictions)
```

```python
f1_rf = f1_score(y_test, x_test_predictions)
```

```python
print(precision_rf, recall_rf, f1_rf)
```

```python
cm_rf = confusion_matrix(y_test, x_test_predictions)
print(cm_rf)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens', square=True,
            xticklabels=['Non-Diabetic', 'Diabetic'],
            yticklabels=['Non-Diabetic', 'Diabetic'])
plt.title('Confusion Matrix for Random Forest Classifier', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
